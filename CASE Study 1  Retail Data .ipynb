## 11. For all customers aged between 25 - 35, find out: a. What was the total amount spent for “Electronics” and “Books” product categories?


```python
Customer_Final.DOB = Customer_Final.DOB.astype(np.Datetime64)
```

```python
customer_Final_copy = Customer_Final.copy()
```

```python
(customer_Final_copy.tran_date - customer_Final_copy.DOB)
```

```python
customer_Final_copy["tran_year"]= [dt.strftime( i, "%Y") for i in customer_Final_copy.tran_date]
```

```python
customer_Final_copy["YOB"]= [dt.strftime( i, "%Y") for i in customer_Final_copy.DOB]
```

```python

customer_Final_copy.tran_year = customer_Final_copy.tran_year.astype("int") 


customer_Final_copy.YOB = customer_Final_copy.YOB.astype("int")
```

```python
customer_Final_copy["Age"] = customer_Final_copy.tran_year-customer_Final_copy.YOB
```

```python
custome_book_ele_age = customer_Final_copy[((customer_Final_copy.Age > 25) | (customer_Final_copy.Age < 35)) & ((customer_Final_copy.prod_cat =="Books")|(customer_Final_copy.prod_cat =="Electronics")) ]
```

```python
custome_book_ele_age.total_amt = custome_book_ele_age.total_amt.astype("float")
```

```python
custome_book_ele_age.total_amt.sum()
```
